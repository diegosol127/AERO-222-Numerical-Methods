```python
import scipy
import matplotlib.pyplot as plt
```

# ODE Solving with `scipy.integrate.solve_ivp()` (RK45)

```python
a = 0
b = 2
y0 = 1
h = 0.1

# t,y = scipy.integrate.solve_ivp(f,[a,b],[y0])
sol = scipy.integrate.solve_ivp(lambda t, y: 1 + t**2 + t*y, [a,b], [y0],max_step=h)

plt.plot(sol.t,sol.y[0])
plt.plot(sol.t,sol.y[0],'*')
plt.show()
```
