```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos
```

# Problem 2

```python
x = np.linspace(0,4,100)
h = x[1] - x[0]

f = 3*cos(5*x) - 2*x**3 + x**2 - 4*x + 16
df_true = -15*sin(5*x) - 6*x**2 + 2*x - 4

df_5pt = np.zeros_like(x)
for i in range(len(x)):
    if i == 0:
        df_5pt[i] = (-25*f[i] + 48*f[i+1] - 36*f[i+2] + 16*f[i+3] - 3*f[i+4]) / (12*h)
    elif i == 1:
        df_5pt[i] = (-3*f[i-1] - 10*f[i] + 18*f[i+1] - 6*f[i+2] + f[i+3]) / (12*h)
    elif i == len(x)-1:
        df_5pt[i] = (3*f[i-4] - 16*f[i-3] + 36*f[i-2] - 48*f[i-1] + 25*f[i]) / (12*h)
    elif i == len(x)-2:
        df_5pt[i] = (-f[i-3] + 6*f[i-2] - 18*f[i-1] + 10*f[i] + 3*f[i+1]) / (12*h)
    else:
        df_5pt[i] = (f[i-2] - 8*f[i-1] + 8*f[i+1] - f[i+2]) / (12*h)

error = abs(df_5pt- df_true)

plt.plot(x,error)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('5-Point Difference Absolute Error')
plt.show()
```

# Problem 3

```python
def f(x):
    return 3*cos(5*x) - 2*x**3 + x**2 - 4*x + 16
def ddf(x):
    return -75*cos(5*x) - 12*x + 2

def ddf_3pt(f,x,h):
    return (f(x-h) - 2*f(x) + f(x+h)) / h**2

x = 3
h1 = 0.1
h2 = 0.01

ddf_true = ddf(x)
df1 = ddf_3pt(f,x,h1)
df2 = ddf_3pt(f,x,h2)

df_RE = df2 + (df2 - df1) / ((h1/h2)**2 - 1)

err1 = abs(df1 - ddf_true)
err2 = abs(df2 - ddf_true)
err3 = abs(df_RE - ddf_true)

print(df1)
print(df2)
print(df_RE)
print(ddf_true)
print()
print(err1)
print(err2)
print(err3)
```

# Problem 4

```python
t = np.array([0,5,10,15,20,25],dtype=float)
T = np.array([80,44.5,30,24.1,21.7,20.7],dtype=float)
h = t[1] - t[0]

dT = np.zeros_like(T,dtype=float)
for i in range(len(T)):
    if i == 0:
        dT[i] = (-3*T[i] + 4*T[i+1] - T[i+2]) / (2*h)
    elif i == len(T)-1:
        dT[i] = (T[i-2] - 4*T[i-1] + 3*T[i]) / (2*h)
    else:
        dT[i] = (T[i+1] - T[i-1]) / (2*h)

plt.plot(t,T)
plt.plot(t,dT)
plt.show()
```
