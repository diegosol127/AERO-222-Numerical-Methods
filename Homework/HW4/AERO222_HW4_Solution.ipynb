```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos
```

```python
x = np.linspace(0,4,100)
h = x[1] - x[0]

f = 3*cos(5*x) - 2*x**3 + x**2 - 4*x + 16
df_true = -15*sin(5*x) - 6*x**2 + 2*x - 4

df_5pt = np.zeros_like(x)
for i in range(len(x)):
    if i == 0:
        df_5pt[i] = (-25*f[i] + 48*f[i+1] - 36*f[i+2] + 16*f[i+3] - 3*f[i+4]) / (12*h)
    elif i == 1:
        df_5pt[i] = (-3*f[i-1] - 10*f[i] + 18*f[i+1] - 6*f[i+2] + f[i+3]) / (12*h)
    elif i == len(x)-1:
        df_5pt[i] = (3*f[i-4] - 16*f[i-3] + 36*f[i-2] - 48*f[i-1] + 25*f[i]) / (12*h)
    elif i == len(x)-2:
        df_5pt[i] = (-f[i-3] + 6*f[i-2] - 18*f[i-1] + 10*f[i] + 3*f[i+1]) / (12*h)
    else:
        df_5pt[i] = (f[i-2] - 8*f[i-1] + 8*f[i+1] - f[i+2]) / (12*h)

error = abs(df_5pt- df_true)
plt.plot(x,error)
plt.show()
```
